```python
# Read the data
data = pd.read_csv('../../../misc/recorded_data/tick_based_100hz_23-02-19.csv')
```

```python
# get first 1000 rows
data = data.iloc[:1000]
```

```python
# List columns
print(data.columns)
```

```python
"""
Index(['tick', 'angle', 'safety_switch_pressed', 'flex_myobrick_pos_encoder',
       'flex_myobrick_torque_encoder', 'flex_myobrick_current',
       'flex_myobrick_pwm', 'flex_myobrick_in_running_state',
       'extend_myobrick_pos_encoder', 'extend_myobrick_torque_encoder',
       'extend_myobrick_current', 'extend_myobrick_pwm',
       'extend_myobrick_in_running_state'],
      dtype='object')
"""
```

```python
# drop boolean columns
data = data.drop(['safety_switch_pressed', 'flex_myobrick_in_running_state', 'extend_myobrick_in_running_state'], axis=1)
```

```python
# normalize data between -1 and 1 except for tick
tick = data['tick']
data = (data - data.min()) / (data.max() - data.min())
data['tick'] = tick

# reverse torque encoder values
data['flex_myobrick_torque_encoder'] = 1 - data['flex_myobrick_torque_encoder']
```

```python
# plot using bokeh
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource

p = figure(width=800, height=600)
p.line(data['tick'], data['angle'], line_width=2, color='red', legend_label="angle")
p.line(data['tick'], data['flex_myobrick_pos_encoder'], line_width=2, color='blue', legend_label="flex_myobrick_pos_encoder")
p.line(data['tick'], data['flex_myobrick_torque_encoder'], line_width=2, color='green', legend_label="flex_myobrick_torque_encoder")
#p.line(data['tick'], data['flex_myobrick_current'], line_width=2, color='orange', legend_label="flex_myobrick_current")
p.line(data['tick'], data['flex_myobrick_pwm'], line_width=2, color='purple', legend_label="flex_myobrick_pwm")


show(p)
# increase the size of the plot
```
